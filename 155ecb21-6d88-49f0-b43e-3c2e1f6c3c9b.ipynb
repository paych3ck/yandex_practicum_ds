## –ú–æ–¥–µ–ª—å

### –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∏—Ç–æ–≥–æ–≤–æ–≥–æ sMAPE.

```python
def sMape(a, f):
    return 1 / len(a) * np.sum(2 * np.abs(f - a) / (np.abs(a) + np.abs(f)) * 100)

def final_sMape(r, f):
    return 0.25 * r + 0.75 * f
```
